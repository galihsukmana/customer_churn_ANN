```python
with open('pipe_prep.pkl','rb') as f:
  pipe_prep = pickle.load(f)

model = tf.keras.models.load_model('model_last/')
```

## Inferencing

```python
df = pd.read_csv('data_inference.csv')
inf = df.reset_index().drop(['index','Unnamed: 0'],axis=1)
inf
```

```python
data_inf_final = pd.DataFrame(pipe_prep.transform(inf))
data_inf_final
```

```python
res_inf = model.predict(data_inf_final)
res_inf
```

```python
res_inf = np.where(res_inf >= 0.5, 1, 0)
res_inf
```
